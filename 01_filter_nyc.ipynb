```python
import pandas as pd
```

reverse mortgage

```python
#read csv

nyc_all_2018 = pd.read_csv('./all_mortgage_data/nyc_all_2018_originated.csv', dtype={
    'county_code': str,
    'census_tract': str
})

nyc_all_2019 = pd.read_csv('./all_mortgage_data/nyc_all_2019_originated.csv', dtype={
    'county_code': str,
    'census_tract': str
})

nyc_all_2020 = pd.read_csv('./all_mortgage_data/nyc_all_2020_originated.csv', dtype={
    'county_code': str,
    'census_tract': str
})

nyc_all_2021 = pd.read_csv('./all_mortgage_data/nyc_all_2021_originated.csv', dtype={
    'county_code': str,
    'census_tract': str
})

nyc_all_2022 = pd.read_csv('./all_mortgage_data/nyc_all_2022_originated.csv', dtype={
    'county_code': str,
    'census_tract': str
})

# filter to reverse mortgage

rm_nyc_all_2018 = nyc_all_2018[(nyc_all_2018['reverse_mortgage'] == 1)]
rm_nyc_all_2019 = nyc_all_2019[(nyc_all_2019['reverse_mortgage'] == 1)]
rm_nyc_all_2020 = nyc_all_2020[(nyc_all_2020['reverse_mortgage'] == 1)]
rm_nyc_all_2021 = nyc_all_2021[(nyc_all_2021['reverse_mortgage'] == 1)]
rm_nyc_all_2022 = nyc_all_2022[(nyc_all_2022['reverse_mortgage'] == 1)]

# concatenate 2018-2022 datasets

rm_nyc_all = pd.concat((rm_nyc_all_2018, rm_nyc_all_2019, rm_nyc_all_2020, rm_nyc_all_2021, rm_nyc_all_2022), axis=0)
```

```python
rm_nyc_all
```

```python
rm_nyc_all.to_csv('./reverse_mortgage_data/rm_nyc_all.csv', index=False)
```

```python
# group by census tract to see mortgage count

nyc_tract_count = rm_nyc_all.groupby('census_tract').agg({
    'lei':'count',
    'tract_population': 'max',
    'tract_minority_population_percent': 'min'
    }).sort_values(by='lei', ascending=False).reset_index()

# rename columns 

nyc_tract_count.columns = ['census_tract', 'rm_no', 'tract_population', 'minority_pct']

# calculate mortgage count per 1000 people

nyc_tract_count['rm_no_per1000'] = nyc_tract_count['rm_no'] / nyc_tract_count['tract_population'] * 1000 
nyc_tract_count.sort_values(by='rm_no', ascending=False).reset_index(drop=True)
```

census

```python
# read census csv

acs2020 = pd.read_csv('./census_data/acs_2020_nyc.csv')
```

```python
acs2020
```

```python
# drop the second header row

acs2020 = acs2020.drop([0])
acs2020
```

```python

# subset to population data

acs_2020_race = acs2020[['GEO_ID', 'NAME', 'DP05_0001E', 'DP05_0071E', 'DP05_0078E']]

# rename columns

acs_2020_race.columns = ['geo_id', 'tract_name', 'population', 'hispanic_pp', 'black_pp']
```

```python
acs_2020_race
```

```python
# split columns

tract_code = acs_2020_race['geo_id'].str.split(pat='US', expand=True)

# insert the census_tract column

acs_2020_race.insert(loc = 1, column='census_tract', value=tract_code[1])

acs_2020_race
```

```python
# correctly type the data 

acs_2020_race = acs_2020_race.astype({
    'population': int,
    'hispanic_pp': int,
    'black_pp': int
})
```

```python
# calculate black and hispanic population percentage

acs_2020_race['hispanic_pp_pct'] = acs_2020_race['hispanic_pp'] / acs_2020_race['population'] * 100
acs_2020_race['black_pp_pct'] = acs_2020_race['black_pp'] / acs_2020_race['population'] * 100

```

```python
acs_2020_race
```

```python
# export

acs_2020_race.to_csv('./census_data/nyc_census_2020.csv', index=False)
```

merge

```python
# merge the census tract info and reverse mortgage info

nyc_rm_census = pd.merge(
    nyc_tract_count,
    acs_2020_race,
    how='left',
    on='census_tract',
    validate='1:1'
)

nyc_rm_census = nyc_rm_census.sort_values(by='rm_no_per1000', ascending=False)

```
