# arizona

## reverse mortgage

```python
#read csv

arizona_all_2021 = pd.read_csv('./all_mortgage_data/arizona_all_2021_originated.csv', dtype={
    'county_code': str,
    'census_tract': str
})

# filter to reverse mortgage

rm_arizona_all_2021 = arizona_all_2021[(arizona_all_2021['reverse_mortgage'] == 1)]


# export to csv

rm_arizona_all_2021.to_csv('./reverse_mortgage_data/south_carolina/rm_arizona_all_2021.csv', index=False)
```

## census

```python
# read census csv

acs2021 = pd.read_csv('./census_data/acs_2021_arizona.csv')

# drop the second header row

acs2021 = acs2021.drop([0])

# subset to population data

acs_2021_race = acs2021[['GEO_ID', 'NAME', 'DP05_0001E', 'DP05_0071E', 'DP05_0078E']]

# rename columns

acs_2021_race.columns = ['geo_id', 'tract_name', 'population', 'hispanic_pp', 'black_pp']

# split geo_id column

tract_code = acs_2021_race['geo_id'].str.split(pat='US', expand=True)

# insert the census_tract column

acs_2021_race.insert(loc = 1, column='census_tract', value=tract_code[1])

# correctly type the data

acs_2021_race = acs_2021_race.astype({
    'population': int,
    'hispanic_pp': int,
    'black_pp': int
})

# calcualate hispanic & black population

acs_2021_race['hispanic_pp_pct'] = acs_2021_race['hispanic_pp'] / acs_2021_race['population'] * 100
acs_2021_race['black_pp_pct'] = acs_2021_race['black_pp'] / acs_2021_race['population'] * 100
```

```python
acs_2021_race
```

```python
# export census table to csv

acs_2021_race.to_csv('./census_data/arizona_census_2021.csv', index=False)
```

## merge

```python
# group by census tract to see mortgage count

arizona_tract_count_2021 = rm_arizona_all_2021.groupby('census_tract').agg({
    'lei':'count',
    'tract_population': 'max',
    'tract_minority_population_percent': 'min'
    }).sort_values(by='lei', ascending=False).reset_index()

# rename columns 

arizona_tract_count_2021.columns = ['census_tract', 'rm_no', 'tract_population', 'minority_pct']

# calculate mortgage count per 1000 people

arizona_tract_count_2021['rm_no_per1000'] = arizona_tract_count_2021['rm_no'] / arizona_tract_count_2021['tract_population'] * 1000 

# sort by count per 1000 people 

arizona_tract_count_2021 = arizona_tract_count_2021.sort_values(by='rm_no_per1000', ascending=False).reset_index(drop=True)
```

```python
# merge the census tract info and reverse mortgage info

arizona_rm_census = pd.merge(
    arizona_tract_count_2021,
    acs_2021_race,
    how='left',
    on='census_tract',
    validate='1:1'
)
```
