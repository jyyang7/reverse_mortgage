```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

## read reverse mortgage data

```python
san_diego_rm_all = pd.read_csv('./reverse_mortgage_data/san_diego_rm_all.csv', dtype={
    'county_code': str,
    'census_tract': str
})
san_diego_rm_all.info()
```

## read census data and filter to county

```python
census_all = pd.read_csv('./census_data/ACS2019_filtered0623.csv', dtype={
    'census_tract': str,
    'county_code': str,
    'census_tract_np': str,
    'census_tract_period': str
})
census_all.info()
```

```python
census_san_diego = census_all[census_all['county_code'] == '06073'].copy()
```

```python
census_san_diego.info()
```

## merge lender list

```python
# read lender list

lender_list_2018 = pd.read_csv('./raw_data/lei_agency_name_2018.csv')
```

```python
lender_list_2018.info()
```

```python
# merge by lei

san_diego_rm_all
```

```python
lender_list_2018
```

```python
san_diego_rm_full = pd.merge(
    san_diego_rm_all,
    lender_list_2018,
    on='lei',
    how='left',
    validate='m:1'
)
```

```python
san_diego_rm_full.info(verbose=True, show_counts=True)
```

```python
san_diego_rm_name = san_diego_rm_full.drop(columns= ['activity_year_y', 'calendar_quarter', 'lar_count']).copy()
```

```python
san_diego_rm_name.head()
```

```python
san_diego_rm_name.info(verbose=True, show_counts=True)
```

## group by census tracts

```python
# filter by action_taken = 1

san_diego_rm_originated = san_diego_rm_name[san_diego_rm_name['action_taken'] == 1]
```

```python
san_diego_rm_originated.info(verbose=True, show_counts=True)
```

```python
# group by census_tract

san_diego_census_counts = san_diego_rm_originated.groupby(['census_tract']).count()
```

```python
san_diego_census_counts = san_diego_census_counts[['lei']].copy()
```

```python
san_diego_census_counts = san_diego_census_counts.reset_index()
```

```python
san_diego_census_counts = san_diego_census_counts.sort_values(by=['lei'], ascending=False).copy()
san_diego_census_counts.rename(columns={'lei': 'rm_no'}, inplace=True)
```

```python
san_diego_census_counts = san_diego_census_counts.reset_index(drop=True)
```

```python
san_diego_census_counts
```

## merge with census data

```python
census_san_diego
```

```python
san_diego_census_rm = pd.merge(
    san_diego_census_counts,
    census_san_diego,
    on='census_tract',
    how='left',
    validate='1:1'
)
```

```python
san_diego_census_rm.sample(10)
```

```python
san_diego_census_rm.info(verbose=True)
```

```python
# subset to the columns that I need

san_diego_census_rm = san_diego_census_rm[['census_tract', 'rm_no', 'county_code', 'county', 'state_code', 'census_tract_np', 'census_tract_period', 'total_population', '62yr_pp_pct', 'median_age', 'hispanic_pp_pct', 'non_hispanic_black_pp_pct']].copy()
```

```python
san_diego_census_rm
```

```python
# calculate rm per 1000 people

san_diego_census_rm['_rm_per_1000'] = (
    san_diego_census_rm['rm_no'] / san_diego_census_rm['total_population'] * 1000
)
```

```python
san_diego_census_rm.info()
```

```python
# get the column names

cols = list(san_diego_census_rm.columns.values)
cols
```

```python
# reorder the columns

san_diego_census_rm = san_diego_census_rm[[
    'census_tract', 
    '_rm_per_1000',
    'rm_no',
    'county_code',
    'county',
    'state_code',
    'census_tract_np',
    'census_tract_period',
    'total_population',
    '62yr_pp_pct',
    'median_age',
    'hispanic_pp_pct',
    'non_hispanic_black_pp_pct',
 ]]
```

```python
# sort by reverse mortgages originated per 1000 people

san_diego_census_rm = san_diego_census_rm.sort_values(by='_rm_per_1000', ascending=False)
san_diego_census_rm
```
