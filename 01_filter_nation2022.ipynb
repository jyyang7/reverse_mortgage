```python
import pandas as pd
```

```python
all_2022 = pd.read_csv('./all_mortgage_data/nation_all_2022.csv', dtype={
    'county_code': str,
    'census_tract': str
})
```

```python
# filter to reverse mortgage
all_rm_2022 = all_2022[(all_2022['reverse_mortgage'] == 1)]
```

```python
all_rm_2022.sample()
```

```python
all_rm_2022.info()
```

```python
# groupby census tract

rm_2022_national = all_rm_2022.groupby(['census_tract']).count()
```

```python
rm_2022_national = rm_2022_national.sort_values(by='lei', ascending=False)
```

```python
rm_2022_national
```

```python
rm_2022_tract_count = rm_2022_national[['lei']].reset_index()
```

```python
rm_2022_tract_count
```

```python
tract_population = all_rm_2022[['census_tract','tract_population', 'tract_minority_population_percent']]
```

```python
tract_population
```

```python
rm_census_2022 = pd.merge(
    rm_2022_tract_count,
    tract_population,
    on='census_tract',
    how='left',
    validate='1:m'
)

rm_census_2022
```

```python
rm_census_2022 = rm_census_2022[['census_tract', 'lei_x', 'tract_population', 'tract_minority_population_percent']].copy()
```

```python
rm_census_2022
```

```python
rm_census_2022['rm_per_1000'] = rm_census_2022['lei_x'] / rm_census_2022['tract_population'] * 1000
rm_census_2022
```

```python
rm_census_2022 = rm_census_2022.drop_duplicates(subset='census_tract',keep='first', inplace=False).reset_index(drop=True)
```

```python
rm_census_2022.head()
```

```python
rm_census_2022.rename(columns={'lei_x': 'rm_no'}, inplace=True)
```

```python
rm_census_2022 = rm_census_2022.sort_values(by='rm_per_1000', ascending=False).reset_index(drop=True)
```
